```python
red = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
white = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
redData = pd.read_csv(red, sep = ";")
whiteData = pd.read_csv(white, sep = ";")
```

```python
redData["type"] = 0
whiteData["type"] = 1
```

```python
wine = whiteData.append(redData, ignore_index=True)
```

```python
wine
```

```python
x = wine.iloc[:, :12].values
y = wine.iloc[:, 12].values
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaller = StandardScaler()
x = scaller.fit_transform(x)
```

```python
trainX, testX, trainY, testY = train_test_split(x,y, test_size = 0.3)
```

```python
trainY
```

```python
#### Deep Learning Part...
```

```python
import tensorflow
tensorflow.__version__
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model = Sequential()
model.add(Dense(20, activation = "relu", input_shape = (12,)))

### OutPut Layer -----
model.add(Dense(1, activation = "sigmoid"))

## Select your Op and Compile Your Model...
model.compile(loss="binary_crossentropy", optimizer = "adam", metrics=["accuracy"])
h = model.fit(trainX, trainY, validation_split=0.3, epochs=30, batch_size = 3000)
```

```python
h.history
```

```python
loss = h.history["loss"]
acc = h.history["accuracy"]
val_loss = h.history["val_loss"]
val_acc = h.history["val_accuracy"]
```

```python
tvA = pd.DataFrame({
    "Ta":acc,
    "Va":val_acc
})
```
